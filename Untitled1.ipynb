# Merge

```python

l = df_2007.loc[:10,['uf','municipio']]
```

```python
r = df_2007.loc[:10,['municipio','tipo_acidente','pessoas']]
```

```python
#apagando os valores duplicados do municipio
l.municipio =l.municipio.drop_duplicates()
r.municipio =r.municipio.drop_duplicates()
```

```python
#apaga as colunas que tem valor NaN
l = l.dropna()
r= r.dropna()
```

```python
#Substiuindo o IGARAPE por LAVRAS
l = l.municipio.replace('IGARAPE','LAVRAS')
```

```python
pd.concat([l,r],axis=1)
```

```python
#Retorna apenas as colunas de l que existem em R
pd.merge(l,r)
```

```python
#Faz a união de todos elementos 
pd.merge(l,r, how='outer',on='municipio')
```

```python
#Retorna o dataset da esquerda e seus valores(Como o muncipio de lavras nao existe no dataframe da direita. as colunas ficaram com o valor nan)
pd.merge(l,r, how='left',on='municipio')
```

```python
# Retorna os valores do dataframe da direita
pd.merge(l,r, how='right',on='municipio')
```
